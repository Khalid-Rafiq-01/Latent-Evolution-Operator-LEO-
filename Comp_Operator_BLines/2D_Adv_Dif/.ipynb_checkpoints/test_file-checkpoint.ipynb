```python
def get_model(latent_dim):
    # Instantiate encoder, decoder, and model
    encoder = Encoder(latent_dim)
    decoder  = Decoder(latent_dim)  # Decoder for x(t)
    propagator = Propagator(latent_dim) # z(t) --> z(t+tau)
    model = Model(encoder, decoder, propagator)
    return model
```

```python
def evaluate_mse_by_zone(
    model,
    exact_solution,
    alpha_interval_split,
    tau_interval_split,
    dt=2.0/500.0,
    num_samples=200,
    max_trials_per_zone=10000,
    t0_range=(0.01, 0.4),
    device="cuda",
):
    model.eval()
    device = torch.device(device)

    alpha_zones = {
        "LE": alpha_interval_split.extrapolation_left,
        "IN": alpha_interval_split.interpolation,
        "RE": alpha_interval_split.extrapolation_right,
    }

    tau_zones = {
        "LE": tau_interval_split.extrapolation_left,
        "IN": tau_interval_split.interpolation,
        "RE": tau_interval_split.extrapolation_right,
    }

    mse_by_zone = {}

    for a_key, a_interval in alpha_zones.items():
        for t_key, t_interval in tau_zones.items():
            zone_name = f"alpha_{a_key}_tau_{t_key}"
            mses = []
            trials = 0

            while len(mses) < num_samples and trials < max_trials_per_zone:
                trials += 1

                alpha_val = float(np.random.uniform(*a_interval))
                tau_val   = float(np.random.uniform(*t_interval))
                t0        = float(np.random.uniform(*t0_range))

                # *** NO MORE t0 + tau * dt < 2 CHECK ***

                x_exact_t0 = exact_solution(alpha_val, t0)
                x_exact_tf = exact_solution(alpha_val, t0 + tau_val*dt)

                x_in = torch.tensor(x_exact_t0, dtype=torch.float32, device=device)[None, None, :, :]
                alpha_in = torch.tensor([[alpha_val]], dtype=torch.float32, device=device)
                tau_in = torch.tensor([[tau_val]], dtype=torch.float32, device=device)

                with torch.no_grad():
                    _, x_hat_tau, *_ = model(x_in, tau_in, alpha_in)

                x_tf_t = torch.tensor(x_exact_tf, dtype=torch.float32, device=device)[None, None, :, :]
                mse = F.mse_loss(x_hat_tau, x_tf_t).item()
                mses.append(mse)

            mean_mse = float(np.mean(mses)) if len(mses) > 0 else float("nan")
            mse_by_zone[zone_name] = {"mean_mse": mean_mse, "mses": mses, "n": len(mses)}

            print(f"{zone_name}: collected {len(mses)} samples (mean MSE = {mean_mse:.3e})")

    return mse_by_zone
```

```python
mse_stats = evaluate_mse_by_zone(
    model=model,
    exact_solution=exact_solution,
    alpha_interval_split=alpha_interval_split,
    tau_interval_split=tau_interval_split,
    dt=dt,
    num_samples=2500,      # e.g. 3000 samples per zone
    t0_range=(0.01, 2.0),
    device="cuda",
)

print(mse_stats["alpha_LE_tau_LE"]["mean_mse"])
```

```python
def plot_zone_mse(mse_by_zone, title="Mean MSE by (alpha, tau) zone"):
    """
    mse_by_zone: dict returned by evaluate_mse_by_zone, with keys like
        'alpha_LE_tau_LE', ..., 'alpha_RE_tau_RE'
        and values {'mean_mse': float, 'mses': [...], 'n': int}
    """

    alpha_order = ["LE", "IN", "RE"]
    tau_order   = ["LE", "IN", "RE"]

    x_positions = []
    x_labels    = []
    y_values    = []

    # Build x positions and labels in 3x3 order
    for i, a_key in enumerate(alpha_order):
        for j, t_key in enumerate(tau_order):
            zone_key = f"alpha_{a_key}_tau_{t_key}"
            if zone_key not in mse_by_zone:
                continue

            mse = mse_by_zone[zone_key]["mean_mse"]
            if (mse is None) or (not np.isfinite(mse)) or (mse <= 0):
                continue

            x = i * 3 + j  # 0..8
            x_positions.append(x)
            y_values.append(mse)

            label = (
                rf"$\alpha_{{\mathrm{{{a_key}}}}}$" + "\n" +
                rf"$\tau_{{\mathrm{{{t_key}}}}}$"
            )
            x_labels.append(label)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_yscale("log")

    if x_positions:
        x_min_idx = min(x_positions)
        x_max_idx = max(x_positions)

        # Axis runs exactly from first zone boundary to last
        left_bound  = x_min_idx - 0.5
        right_bound = x_max_idx + 0.5
        ax.set_xlim(left_bound, right_bound)

        # Light vertical boundaries at every zone edge
        for k in range(x_min_idx, x_max_idx + 2):
            x_b = k - 0.5
            ax.axvline(x_b, color="0.9", linestyle="-", linewidth=0.7, zorder=0)

        # Slightly darker separators between alpha groups
        for x_sep in [2.5, 5.5]:
            ax.axvline(x_sep, color="0.7", linestyle="-", linewidth=1.0, zorder=1)

    # Horizontal segments spanning full zone width, no markers
    half_width = 0.5
    for x, y in zip(x_positions, y_values):
        ax.plot(
            [x - half_width, x + half_width],
            [y, y],
            linewidth=3.0,
            color="C0",
            solid_capstyle="butt",
            zorder=5,   # ensure above the grid/vertical lines
        )

    ax.set_xticks(x_positions)
    ax.set_xticklabels(x_labels, fontsize=12)

    ax.set_ylabel("Mean MSE (log scale)", fontsize=14)
    ax.set_xlabel(r"$(\alpha, \tau)$ zones", fontsize=14)
    ax.set_title(title, fontsize=14)

    ax.grid(True, which="both", axis="y", alpha=0.2)

    plt.tight_layout()
    return fig, ax
```

```python
mse_by_zone = evaluate_mse_by_zone(
    model,
    exact_solution,
    alpha_interval_split,
    tau_interval_split,
    dt=2.0/500.0,
    num_samples=1000,
    device="cuda",
)

fig, ax = plot_zone_mse(mse_by_zone, title="Direct model: zone-wise MSE")
plt.show()
```

## Mass Conservation

```python
# Mass conservation - 
def compute_mass(U, L=4.0):
    """
    Computes the "mass" of a batch of 2D fields on a domain [-L, L] x [-L, L].
    
    Arguments:
        U:  A PyTorch tensor of shape (batch_size, 1, Nx, Ny).
            We assume it's the 2D field values you want to integrate.
        L:  Domain half-length in each dimension (default=2.0).
    
    Returns:
        mass: A 1D tensor of length batch_size, giving the integral of U
              over the domain for each sample in the batch.
    """
    # Extract the spatial resolution
    Nx = U.size(-2)  # e.g. 128
    Ny = U.size(-1)  # e.g. 128
    
    # Compute the grid spacing in each dimension
    dx = (2.0 * L) / Nx
    dy = (2.0 * L) / Ny
    
    # Sum over channel, height, width => dims (1, 2, 3).
    # shape of U is (batch, channel=1, Nx, Ny).
    # The result is shape (batch,) for each sample.
    mass = U.sum(dim=[1, 2, 3]) * (dx * dy)
    
    return mass
```

```python
def evaluate_mass_error_by_zone(
    model,
    exact_solution,
    alpha_interval_split,
    tau_interval_split,
    dt=2.0/500.0,
    num_samples=200,
    max_trials_per_zone=10000,
    t0_range=(0.0, 1.2),
    device="cuda",
    eps=1e-10,
):
    """
    Compute mass statistics for all 9 (alpha, tau) zones.

    For each zone (alpha_zone in {LE, IN, RE}, tau_zone in {LE, IN, RE}):
      - Sample (alpha, tau, t0) as in evaluate_mse_by_zone
      - Keep only samples with t0 + tau * dt < 2
      - Compute mass of exact and predicted field at t0 + tau*dt
      - Record:
          mass_ratio = M_pred / (M_exact + eps)
          rel_err    = |mass_ratio - 1|

    Returns
    -------
    mass_err_by_zone : dict
        Keys: "alpha_LE_tau_LE", ..., "alpha_RE_tau_RE"
        Values:
          {
            "mean_mass_ratio": float,
            "mass_ratios": list_of_floats,
            "mean_rel_err": float,
            "rel_errs": list_of_floats,
            "n": int
          }
    """

    model.eval()
    device = torch.device(device)

    alpha_zones = {
        "LE": alpha_interval_split.extrapolation_left,
        "IN": alpha_interval_split.interpolation,
        "RE": alpha_interval_split.extrapolation_right,
    }

    tau_zones = {
        "LE": tau_interval_split.extrapolation_left,
        "IN": tau_interval_split.interpolation,
        "RE": tau_interval_split.extrapolation_right,
    }

    mass_err_by_zone = {}

    for a_key, a_interval in alpha_zones.items():
        for t_key, t_interval in tau_zones.items():
            zone_name = f"alpha_{a_key}_tau_{t_key}"
            mass_ratios = []
            rel_errs    = []
            trials = 0

            while len(mass_ratios) < num_samples and trials < max_trials_per_zone:
                trials += 1

                # Sample alpha, tau, t0
                alpha_val = float(np.random.uniform(*a_interval))
                tau_val   = float(np.random.uniform(*t_interval))
                t0        = float(np.random.uniform(*t0_range))

                # Exact fields
                x_exact_t0 = exact_solution(alpha_val, t0)              # (H, W)
                x_exact_tf = exact_solution(alpha_val, t0 + tau_val*dt) # (H, W)

                # Tensor inputs for model
                x_in = torch.tensor(
                    x_exact_t0, dtype=torch.float32, device=device
                )[None, None, :, :]  # (1,1,H,W)

                alpha_in = torch.tensor(
                    [[alpha_val]], dtype=torch.float32, device=device
                )  # (1,1)

                tau_in = torch.tensor(
                    [[tau_val]], dtype=torch.float32, device=device
                )  # (1,1)

                with torch.no_grad():
                    _, x_hat_tau, *_ = model(x_in, tau_in, alpha_in)

                # Exact tf as tensor
                x_tf_t = torch.tensor(
                    x_exact_tf, dtype=torch.float32, device=device
                )[None, None, :, :]  # (1,1,H,W)

                # Compute masses
                M_exact = compute_mass(x_tf_t).item()
                M_pred  = compute_mass(x_hat_tau).item()

                # Mass ratio and relative error
                mass_ratio = M_pred / (M_exact + eps)
                rel_err    = abs(mass_ratio - 1.0)

                mass_ratios.append(mass_ratio)
                rel_errs.append(rel_err)

            mean_mass_ratio = float(np.mean(mass_ratios)) if len(mass_ratios) > 0 else float("nan")
            mean_rel_err    = float(np.mean(rel_errs))    if len(rel_errs) > 0 else float("nan")

            mass_err_by_zone[zone_name] = {
                "mean_mass_ratio": mean_mass_ratio,
                "mass_ratios": mass_ratios,
                "mean_rel_err": mean_rel_err,
                "rel_errs": rel_errs,
                "n": len(mass_ratios),
            }

            print(
                f"{zone_name}: collected {len(mass_ratios)} samples "
                f"(mean mass ratio = {mean_mass_ratio:.3e}, "
                f"mean rel err = {mean_rel_err:.3e})"
            )

    return mass_err_by_zone


model = model.cuda()
mass_stats = evaluate_mass_error_by_zone(
    model=model,
    exact_solution=exact_solution,
    alpha_interval_split=alpha_interval_split,
    tau_interval_split=tau_interval_split,
    dt=dt,
    num_samples=3000,
    t0_range=(0.0, 2.0),
    device="cuda",
)

print(mass_stats["alpha_LE_tau_LE"]["mean_mass_ratio"])
print(mass_stats["alpha_LE_tau_LE"]["mean_rel_err"])



def plot_zone_mass_ratio(mass_err_by_zone, title="Mean mass ratio by (alpha, tau) zone"):
    alpha_order = ["LE", "IN", "RE"]
    tau_order   = ["LE", "IN", "RE"]

    x_positions = []
    x_labels    = []
    y_values    = []

    for i, a_key in enumerate(alpha_order):
        for j, t_key in enumerate(tau_order):
            zone_key = f"alpha_{a_key}_tau_{t_key}"
            if zone_key not in mass_err_by_zone:
                continue

            ratio = mass_err_by_zone[zone_key]["mean_mass_ratio"]
            if (ratio is None) or (not np.isfinite(ratio)):
                continue

            x = i * 3 + j
            x_positions.append(x)
            y_values.append(ratio)

            label = (
                rf"$\alpha_{{\mathrm{{{a_key}}}}}$" + "\n" +
                rf"$\tau_{{\mathrm{{{t_key}}}}}$"
            )
            x_labels.append(label)

    fig, ax = plt.subplots(figsize=(10, 5))

    if x_positions:
        x_min_idx = min(x_positions)
        x_max_idx = max(x_positions)
        left_bound  = x_min_idx - 0.5
        right_bound = x_max_idx + 0.5
        ax.set_xlim(left_bound, right_bound)

        # Vertical zone boundaries
        for k in range(x_min_idx, x_max_idx + 2):
            x_b = k - 0.5
            ax.axvline(x_b, color="0.9", linestyle="-", linewidth=0.7, zorder=0)

        # Separators between alpha groups
        for x_sep in [2.5, 5.5]:
            ax.axvline(x_sep, color="0.7", linestyle="-", linewidth=1.0, zorder=1)

    # Horizontal segments (zone-wise mean ratio)
    half_width = 0.5
    for x, y in zip(x_positions, y_values):
        ax.plot(
            [x - half_width, x + half_width],
            [y, y],
            linewidth=3.0,
            color="C1",
            solid_capstyle="butt",
            zorder=5,
        )

    # Reference line at perfect conservation
    ax.axhline(1.0, color="k", linestyle="--", linewidth=1.0, zorder=4)

    # Optional: tighten y-limits around 1
    if y_values:
        ymin = min(y_values)
        ymax = max(y_values)
        pad  = 0.02 * max(1.0, ymax - ymin)
        ax.set_ylim(ymin - pad, ymax + pad)

    ax.set_xticks(x_positions)
    ax.set_xticklabels(x_labels, fontsize=12)

    ax.set_ylabel(r"Mean mass ratio $M_{\mathrm{pred}} / M_{\mathrm{exact}}$", fontsize=14)
    ax.set_xlabel(r"$(\alpha, \tau)$ zones", fontsize=14)
    ax.set_title(title, fontsize=14)
    ax.grid(True, which="both", axis="y", alpha=0.2)

    plt.tight_layout()
    return fig, ax
```

```python
fig, ax = plot_zone_mass_ratio(mass_stats,
                               title="Direct model: zone-wise mass ratio")
plt.show()
```

```python
def plot_zone_mass_ratio(
    mass_err_by_zone,
    title="Mean mass ratio by (alpha, tau) zone",
    savepath="adv_diff_mass_ratio.png",
):
    alpha_order = ["LE", "IN", "RE"]
    tau_order   = ["LE", "IN", "RE"]

    x_positions = []
    x_labels    = []
    y_values    = []

    for i, a_key in enumerate(alpha_order):
        for j, t_key in enumerate(tau_order):
            zone_key = f"alpha_{a_key}_tau_{t_key}"
            if zone_key not in mass_err_by_zone:
                continue

            ratio = mass_err_by_zone[zone_key]["mean_mass_ratio"]
            if (ratio is None) or (not np.isfinite(ratio)):
                continue

            x = i * 3 + j
            x_positions.append(x)
            y_values.append(ratio)

            label = (
                rf"$\alpha_{{\mathrm{{{a_key}}}}}$" + "\n" +
                rf"$\tau_{{\mathrm{{{t_key}}}}}$"
            )
            x_labels.append(label)

    fig, ax = plt.subplots(figsize=(10, 5))

    if x_positions:
        x_min_idx = min(x_positions)
        x_max_idx = max(x_positions)
        left_bound  = x_min_idx - 0.5
        right_bound = x_max_idx + 0.5
        ax.set_xlim(left_bound, right_bound)

        # Vertical zone boundaries
        for k in range(x_min_idx, x_max_idx + 2):
            x_b = k - 0.5
            ax.axvline(x_b, color="0.9", linestyle="-", linewidth=0.7, zorder=0)

        # Separators between alpha groups
        for x_sep in [2.5, 5.5]:
            ax.axvline(x_sep, color="0.7", linestyle="-", linewidth=1.0, zorder=1)

    # Horizontal segments (zone-wise mean ratio)
    half_width = 0.5
    for x, y in zip(x_positions, y_values):
        ax.plot(
            [x - half_width, x + half_width],
            [y, y],
            linewidth=3.0,
            color="C0",              # blue
            solid_capstyle="butt",
            zorder=5,
        )

    # Reference line at perfect conservation
    ax.axhline(1.0, color="k", linestyle="--", linewidth=1.0, zorder=4)

    # Tighten y-limits around 1
    if y_values:
        ymin = min(y_values)
        ymax = max(y_values)
        pad  = 0.02 * max(1.0, ymax - ymin)
        ax.set_ylim(ymin - pad, ymax + pad)

    ax.set_xticks(x_positions)
    ax.set_xticklabels(x_labels, fontsize=12)

    ax.set_ylabel(r"Mean mass ratio $M_{\mathrm{pred}} / M_{\mathrm{exact}}$", fontsize=14)
    ax.set_xlabel(r"$(\alpha, \tau)$ zones", fontsize=14)
    ax.set_title(title, fontsize=14)
    ax.grid(True, which="both", axis="y", alpha=0.2)

    plt.tight_layout()

    # save figure
    if savepath is not None:
        fig.savefig(savepath, dpi=300, bbox_inches="tight")

    return fig, ax
```
